```python
# BFS - Breadth First Search
tree = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': [],
    'F': []
}
```

```python
def bfs(tree, start):
  visited = []
  queue = [start]

  while queue:
    node=queue.pop(0)
    visited.append(node)
    queue.extend(tree[node])
  return visited
```

```python
bfs(tree, 'A')
```

```python
# DFS - Depth First Search
tree = {
    'A':['B','C'],
    'B':['D','E', 'F'],
    'C':['G'],
    'D':[],
    'E':['H'],
    'F':[],
    'G':[],
    'H':[]
}
```

```python
def dfs(tree, node, visited=None):
  if visited is None:
    visited=[]

  visited.append(node)
  for child in tree[node]:
    dfs(tree, child, visited)
  return visited
```

```python
print('DFS:', dfs(tree,'A'))
```
