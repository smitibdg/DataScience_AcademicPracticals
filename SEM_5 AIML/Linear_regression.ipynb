```python
# Simple Linear Regression OR Univariate Linear Regression
# Importing required libraries
import numpy as np # For numerical operations and reshaping
import pandas as pd # for loading and data manipulation
import matplotlib.pyplot as plt # for plotting graphs
import seaborn as sns # for advanced statistical
from sklearn.model_selection import train_test_split # To split data into training and testing
from sklearn.linear_model import LinearRegression # To create the regression model
from sklearn.metrics import r2_score, mean_squared_error # To evaluate model performance
```

```python
df = pd.read_csv('/content/housingData-Real.csv')
```

```python
#feature extraction
df.drop(['id','date'], axis=1, inplace=True)
```

```python
df.head()
```

```python
x = df.sqft_living # Independent V
y = df.price # Dependent V
# as surface area increases the price of room will also increase
```

```python
X = np.array(x).reshape(-1,1)
Y = np.array(y).reshape(-1,1)
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X, Y, test_size = 0.2, random_state=101)
```

```python
model1 = LinearRegression()
model1.fit(X_train, Y_train)
```

```python
Y_pred = model1.predict(X_test)
a = r2_score(Y_test, Y_pred)
a
```

```python
# Plotting of training data model
plt.scatter(X_train, Y_train)
plt.plot(X_train, model1.predict(X_train), color='red')
```

```python
# Plotting of testing data model
plt.scatter(X_test, Y_test)
plt.plot(X_train, model1.predict(X_train), color='red')
```

```python
x = df.drop(['price'], axis=1) # make all column independent variable except price column
y = df.price # D.V
```

```python
x_train,x_test,y_train,y_test = train_test_split(x, y, random_state=101)
```

```python
model2 = LinearRegression()
model2.fit(x_train, y_train)
```

```python
y_pred = model2.predict(x_test)
```

```python
b = r2_score(y_test, y_pred)
b # Error Rate
```

```python
print("r2 score of the Univariate Linear Regression is : {}".format(a))
print("r2 score of the Multiple Linear Regression is : {}".format(b))
```
